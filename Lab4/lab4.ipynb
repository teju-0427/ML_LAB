```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])   #no. of hours studied
y = np.array([40, 50, 60, 70, 80])   #marks obtained
n = len(x)  #no.of students
```

```python
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x * x)
```

```python
m = (n * sum_xy - sum_x * sum_y)/(n * sum_x2 - sum_x * sum_x)
c = (sum_y-m*sum_x)/n
```

```python
y_pred = m*x+c
```

```python
print(f"Slope(m) = {m:.2f}")
print(f"Intercept(c) = {c:.2f}")
print("Predicted marks = ", np.round(y_pred).astype(int))
```

```python
plt.scatter(x,y,color = 'blue', label = "Students")
plt.plot(x,y_pred,color = 'red', label = "Regression Line")
plt.legend()
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```
